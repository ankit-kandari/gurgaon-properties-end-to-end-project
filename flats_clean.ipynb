```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("flats.csv")
```

```python
pd.set_option("display.max_rows",None)
```

```python
df.head()
```

```python
df = df.drop(columns = ["link"])
```

```python
df["society"]=df["society"].str.lower()
```

```python
len(df["society"].value_counts())
```

```python
import re
```

```python
####.value_counts
#df["society"].apply(lambda x: re.sub(r"\d+\.\d+\s?★$","",str(x)))
#len(df["society"].apply(lambda x: re.sub(r'\d+(\.\d+)?\s?★',"",str(x)).strip()
df["society"]=df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df.head()
```

```python
df["price"].value_counts()
```

```python
df = df[~(df["price"]=="Price on Request")]
```

```python
def price_con(val):
    if type(val)==float:
        return val
    elif isinstance(val, list) and len(val) >= 2:
        if val[1] == "Lac":
            return round(float(val[0])/100,2)
        else:
            return round(float(val[0]),2)
    
```

```python
df["price"]=df["price"].str.split(" ").apply(price_con)
```

```python
df.head()
```

```python
df["area"]=df["area"].str.split("/").str.get(0).str.replace("₹","").str.replace(",","").str.strip()
```

```python
df.drop(286,inplace = True)
```

```python
df["area"]=df["area"].astype("float")
```

```python
df=df.rename(columns = {"area":"persf_area","bedRoom":"bedroom"})
```

```python
df.head()
```

```python
df["bedroom"]=df["bedroom"].str.split().str.get(0).str.strip().astype("float")
```

```python
df[df.bedroom.isnull()].index
```

```python
df = df.drop(df[df.bedroom.isnull()].index)
```

```python
df["bathroom"]=df["bathroom"].str.split().str.get(0).str.strip().astype("float")
```

```python
df["balcony"]=df["balcony"].str.split().str.get(0).str.replace("No","").str.strip()
```
