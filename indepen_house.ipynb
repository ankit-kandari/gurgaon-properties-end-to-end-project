```python
df = df.drop_duplicates()
```

```python
df=df.rename(columns = {"rate":"persf_area","bedRoom":"bedroom"})
```

```python
print(len(df["society"].value_counts()))
print("==============================================")
df["society"].value_counts()
```

```python
import re
```

```python
df['society'] = df['society'].apply(lambda x: re.sub(r'\d+(\.\d+)?\s?★', '', str(x)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def price_con(val):
    if type(val)==float:
        return val
    elif isinstance(val, list) and len(val) >= 2:
        if val[1] == "Lac":
            return round(float(val[0])/100,2)
        else:
            return round(float(val[0]),2)
```

```python
df["price"]=df["price"].str.split(" ").apply(price_con)
```

```python
df.head(3)
```

```python
df["persf_area"]=df["persf_area"].str.split("/").str.get(0).str.replace("₹","").str.replace(",","").str.strip()
```

```python
df["persf_area"]=df["persf_area"].astype(float)
```

```python
df[df['bedroom'].isnull()]
```

```python
df = df[~df['bedroom'].isnull()]
```

```python
df.shape
```

```python
df['bedroom'] = df['bedroom'].str.split(' ').str.get(0).astype('int')
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.head()
```

```python
df['noOfFloor'].value_counts()
```

```python
df['noOfFloor'].isnull().sum()
```

```python
df['noOfFloor'] = df['noOfFloor'].str.split(' ').str.get(0)
```

```python
df.head()
```

```python
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df['area'] = round((df['price']*10000000)/df['persf_area'])
```
